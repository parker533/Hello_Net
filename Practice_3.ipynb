There is a difference between normal/gaussian distribution and binomial distribution. The difference is that normal is a form of contineous distribution whilst binomial is a descrete distribution but if the data is large, they look similar.


```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
#generating the normal
sns.kdeplot(random.normal(loc=50, scale=5, size=1000),label = "normal")
#generating the binomial
sns.kdeplot(random.binomial(n=100, p=0.5, size=1000), label = "binomial")
plt.show()
```

```python
# Observing Poisson distribution
sns.displot(random.poisson(lam = 2, size = 1000))
plt.show()
```

```python
# Uniform distribution --> A distribution in which all numbers have the same chance of occuring
# By default it gives numbers between 0 and 1
import matplotlib.pyplot as plt
from numpy import random
import seaborn as sns
sns.kdeplot(random.uniform(size=1000))
plt.show()
```

```python
#multinomial distribution is just a general version of binomimal
# Also, in multidimensional,n is number of outcomes but in binomial it is number of trials.
# Has no size so number of results is = to the n number
# amount of p should match the n number
#It is pvals not p over here
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
sns.kdeplot(random.multinomial(n= 6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6]))
plt.show()
```

**Exponential distribution being the Opposite of poisson distribution**
Poisson distribution measures the amount of occurance in a particular time period eg like How many times do you normally eat in a day which for you is 2. Poisson represents this with lam =. Exponential is the opposite it looks at the time gaps between the occurance, so in this instant there is 1 time gap between the two meals. It represents this by scale =.

```python
#Exponential distribution
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
sns.kdeplot(random.exponential(scale = 2, size = 100))
plt.show()
```

**Chi square distribution and degrees of freedom**
Chi square is used to verify hypothesis. It takes a degree of freedom(df) parameter.
Degree of freedom is pretty much amount of data left after analysis that gives you a wiggle room. eg If you have 2 cordinates and you draw a line with it. you cannot do much statistical stuff because the points are 2 so you have 0 df. another example is at work you work 40hrs a week so you can flex the time which is the df, if you do 30hrs by thursday it means you have to do 10hrs on friday because there is no more wiggle room so there is no more df.


```python
#chisquare
sns.kdeplot(random.chisquare(df=2, size=1000))
plt.show()
```

```python
#Rayleigh distribution is used to analyze signals
# Its scale = is std
#At 2df of chisquare,its figure is similar to releigh
sns.kdeplot(random.rayleigh(scale=2, size=1000))
plt.show()
```

```python
#Pareto distribution
# It is based on the pareto 20%-80% rule
# It takes an a= which is known as a shape parameter
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.histplot(random.pareto(a=5, size=1000))
plt.show()
```

```python
#zipf distribution --> It is based on the zipf law
# It takes an a= which is a distribution parameter
# eg The law says that if a word is the 5th used then it appears 1/5 times compared to the most used.
x = random.zipf(a=2, size=((4,3))) # To show that you can make an array with all the random stuff
print(x)
y = random.zipf(a=3,size=1000)
sns.displot(y[y<=25])
# condition above tells you that in the values in the 1000, they have to be <=25
plt.show()
```
